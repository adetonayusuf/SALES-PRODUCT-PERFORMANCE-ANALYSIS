#### Products that are sold together

```python
# Products that are sold together

Products_Sold_Together = superstore[superstore['order_id'].duplicated(keep=False)]

Products_Sold_Together
```

```python
Products_Sold_Together = superstore[superstore['order_id'].duplicated(keep=False)]

Products_Sold_Together['Grouped'] = superstore.groupby('order_id')['product_name'].transform(lambda x: ','.join(x))

Products_Sold_Together.head()
```

```python
Products_Sold_Together = superstore[superstore['order_id'].duplicated(keep=False)]

Products_Sold_Together['Grouped'] = superstore.groupby('order_id')['product_name'].transform(lambda x: ','.join(x))

Products_Sold_Together = Products_Sold_Together[['order_id', 'Grouped']].drop_duplicates()

Products_Sold_Together.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in Products_Sold_Together['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)

# Common 2
```

These are the most 2 products sold together. It'll be advisable to advertise them together, in order to generate more income from them.

```python
count = Counter()

for row in Products_Sold_Together['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))

for key, value in count.most_common(10):
    print(key, value)

# Common 3
```

These are the most 3 products sold together. It'll be advisable to advertise them together, in order to generate more income from them if possible.

```python
count = Counter()

for row in Products_Sold_Together['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 4)))

for key, value in count.most_common(10):
    print(key, value)

# Common 4
```

```python
These are the most 4 products sold together. It'll be advisable to advertise them together, in order to generate more income from them if possible.
```
